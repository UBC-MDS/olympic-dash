```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("../data/raw/olympics_data.csv", index_col = 0)
```

```python
df.describe()
```

```python
# Filters
# Olympics Type
summer = 1
winter = 1

# Medals
bronze = 1
silver = 1
gold = 1
```

```python
# Filter for Olympics Type
filter_df = df
```

```python
# Filter for Medals
filter_df = filter_df[filter_df['medal'].notna()]
```

```python
filter_df.head()
```

```python
year = 2000
```

```python
temp = filter_df[filter_df['year'] == year]

df1 = pd.DataFrame()

df1['athletes'] = temp.groupby(['noc'])['id'].nunique()
df1['metal_count'] = temp.groupby(['noc'])['medal'].count()
df1['ave_metals'] = df1['metal_count'] / df1['athletes']
```

```python
df1.head()
```

```python
df1 = df1.reset_index()
```

```python
chart = alt.Chart(df1).mark_circle().encode(
                y = 'athletes',
                x = 'ave_metals',
                size = 'metal_count',
                tooltip='noc').interactive()

chart
```
