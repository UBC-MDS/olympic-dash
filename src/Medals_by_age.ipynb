```python
order = ["Gold", "Silver", "Bronze"]
# plot_df = filter_df

plot_df = filter_df[filter_df['age'].between(10, 50)]
plot_df["order"] = filter_df["medal"].replace({ 'Bronze' : 1, 'Silver' : 2, 'Gold' : 3 })

alt.Chart(plot_df).mark_area(
    interpolate='step'
).encode(
    alt.X('age:Q', bin=alt.Bin(maxbins=100)),
    alt.Y('count()'),
    alt.Color('medal:N',
              sort=["Gold", "Silver", "Bronze"],
              scale=alt.Scale(
                        domain=['Bronze', 'Silver', 'Gold'],
                        range=['#CD7F32', '#C0C0C0', '#FFD700'])),
    order=alt.Order('order', sort='ascending')
).properties(
    title='Overlapping Histograms from Tidy/Long Data'
)
```
